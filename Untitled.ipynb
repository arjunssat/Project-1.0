```python
#Titanic Analysis
import numpy as np
import pandas as pd
data=pd.read_csv('D:/Data Analysis/titanic/tested.csv',sep=',')
data.head(3)
```

```python
data.describe()
```

```python
data.isna().sum() # Checking Missing values
```

```python
# Data cleaning
x=data['Age'].mean()
data['Age'].fillna(x,inplace=True)
y=data['Fare'].mean()
data['Fare'].fillna(y,inplace=True)
z=data['Cabin'].mode()[0]
data['Cabin'].fillna(z,inplace=True)
```

```python
data.isna().sum() 
```

```python
data.Survived.mean()
```

```python
data.groupby(['Sex','Pclass']).Survived.mean()
```

```python
data.groupby(['Sex','Pclass']).Survived.mean().unstack()
```

```python
#Pivot table
data.pivot_table(index='Sex',columns='Pclass',values='Survived',aggfunc='mean',margins=True)
```

```python
#No: of records in each categories
data.pivot_table(index='Sex',columns='Pclass',values='Survived',aggfunc='count',margins=True)
```

```python
#Converting continous data into categorical data
pd.cut(data.Age,bins=[0,18,25,99],labels=['child','teen','adult']).head(10)
```
